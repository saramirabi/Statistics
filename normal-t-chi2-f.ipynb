```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 0 #Mean
sigma = 1 #Standard Deviation

num_samples = 10000
samples = np.random.normal(mu, sigma, num_samples)

#print(samples)
```

```python
plt.hist(samples, bins= 30, density= True, alpha= 0.6, color= 'blue')
plt.title('Histogram of generated samples', fontsize = '10')
plt.xlabel('Values of random variable x:', fontsize = '10')
plt.ylabel('Density of variable x:', fontsize= '10')

x_range = np.linspace(-4,4,1000)
pdf_values = norm.pdf(x_range, mu, sigma)

plt.plot(x_range, pdf_values, 'r-', linewidth=2)
```

```python
mean = 0 
std_dev = 1
quantile = 1.96

cumulative_prob = norm.cdf(quantile, loc = mean, scale= std_dev)
probability_greater_than = 1- cumulative_prob

print(f"Probability lower than {quantile}: {cumulative_prob:.2f}")
print(f"Probability greater than {quantile}: {probability_greater_than:.2f}")
```

```python
mean = 0 
std_dev = 1

cumulative_prob = 0.95
#inverse PCF -> Percent Point Function
quantile = norm.ppf(cumulative_prob, loc= mean, scale = std_dev)

print(f"Quantile of Cumulative Probability: {cumulative_prob:.2f}:{quantile:.2f}")
```

```python
mean = 0 
std_dev = 1

lower_quantile = -1.0
upper_quantile = 1.0

lower_prob = norm.cdf(lower_quantile, loc = mean, scale = std_dev)
upper_prob = norm.cdf(upper_quantile, loc=mean, scale= std_dev)

probability_between_quantiles = upper_prob - lower_prob

print(f"Probability between {lower_quantile:.2f} and {upper_quantile:.2f}: {probability_between_quantiles:.2f}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import t
```

```python
degrees_of_freedom = 10 

samples = np.random.standard_t(degrees_of_freedom, size =2000)

print(samples)
```

```python
plt.title('histogram of generated samples', fontsize= '10')
plt.xlabel('Values of random variables x', fontsize= '10')
plt.ylabel('Density of variable x', fontsize= '10')
plt.hist(samples, bins= 30, density= True, alpha= 0.6, color='blue')

x_range = np.linspace(-5, 5, 1000)
pdf_values = t.pdf(x_range, degrees_of_freedom)

plt.plot(x_range, pdf_values, 'r-', linewidth = 2)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm

mean = 0
std_dev = 1

degrees_of_freedom = 50

plt.title("Normal(red) versus t (blue) distribution", fontsize = '10')
plt.xlabel("Values of random variable x", fontsize='10')
plt.ylabel("Density of variable x", fontsize = '10')

x_range = np.linspace(-5, 5, 1000)
pdf_values = norm.pdf(x_range, mean, std_dev)
plt.plot(x_range, pdf_values, 'r-', linewidth = 2)

pdf_values = t.pdf(x_range, degrees_of_freedom)

plt.plot(x_range, pdf_values, 'b-', linewidth = 2)


```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import chi2

degrees_of_freedom = 5

samples = np.random.chisquare(degrees_of_freedom, size =1000)
print(samples)
```

```python
plt.title(f"Chi-square histogram df={degrees_of_freedom}", fontsize = '10')
plt.xlabel('Values of random variable x', fontsize ='10')
plt.ylabel('Density of variable x', fontsize='10')
plt.hist(samples, bins=30, density = True, alpha= 0.6, color= 'blue', label='Chi-Square Samples')

x_range = np.linspace(0, max(samples), 1000)

pdf_values = chi2.pdf(x_range, degrees_of_freedom)

plt.plot(x_range, pdf_values,'r-', linewidth= 2, label='PDF')

plt.legend()
```

```python
degrees_of_freedom = 5

quantile = 10

cumulative_prob = chi2.cdf(quantile, degrees_of_freedom)
Probability_greater_than = 1-cumulative_prob

print(f"Probability lower than {quantile}: {cumulative_prob:.2f}")
print(f"Probability greater than {quantile}: {Probability_greater_than:.2f}")
```

```python
degrees_of_freedom = 5

cumulative_prob = 0.95

quantile = chi2.ppf(cumulative_prob, degrees_of_freedom)

print(f"Quantile for cumulative probability: {cumulative_prob:.2f}: {quantile:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f

dfn = 5
dfd = 10

sample_size = 1000

samples = np.random.f(dfn, dfd, size = sample_size)

#print(samples)
```

```python
plt.hist(samples, bins = 20, density= True, alpha = 0.6, color='b', label='F-Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f"F-Distribution Histogram (dfn={dfn}, dfd={dfd})")

x = np.linspace(0, max(samples), 1000)
pdf = f.pdf(x, dfn, dfd)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

plt.legend()
plt.show()
```
