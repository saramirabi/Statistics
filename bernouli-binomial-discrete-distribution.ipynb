```python
### The probability mass function of the Bernouli distribution is defined as below:
### P(X=1) = p success
### P(X=0) = 1-p failure
# Coin Flips # Binary Classification
#Generate Random Samples -> Assess sample disctibution -> Compare to the population distribution

#Generate Random Samples
import numpy as np  #For generating random data
p = 0.5 #probability
num_samples = 1000 
# we have N 
samples = np.random.binomial(1, p , num_samples)
print(samples)
```

```python
# Calculate Frequency Table
unique_elements, counts = np.unique(samples, return_counts = True)
percentages = (counts/num_samples)* 100
value_counts_and_percentages = zip(unique_elements, counts, percentages)
for i ,j ,k in value_counts_and_percentages:
    print(f"{i}: Count: {j}, Percentage:{k:.2f}%")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli 

bd = bernoulli(0.5)
x=[0,1]
fig, axs = plt.subplots(1, 2, figsize=(10,5))

axs[0].bar(unique_elements, percentages, color = 'blue')
axs[0].set_xlim(-1,2)
axs[0].set_ylim(0,100)
axs[0].set_title("Bernouli distribution(p=0.2) - Sample", fontsize= '10')
axs[0].set_xlabel('Values of random variable x (0,1)', fontsize= '10')
axs[0].set_ylabel('percentage', fontsize='10')

axs[1].bar(x, bd.pmf(x), color='green')
axs[1].set_xlim(-1,2)
axs[1].set_ylim(0,1)
axs[1].set_title("Bernouli distribution (p=0.2) - Population", fontsize='10')
axs[1].set_xlabel("Values of random variable x (0,1)", fontsize= '10')
axs[1].set_ylabel('Probablity', fontsize='10')
```

```python
#Binomial Distribution
#PMF = Probability Mas Function

import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import binom

n = 5 #Number of trials
p = 0.4 #Probability of success
num_samples = 10 # Number of Samples
samples = np.random.binomial(n, p, num_samples)
print(samples)
```

```python
#Frequency Table

unique_elements, counts = np.unique(samples, return_counts= True)
percentages = (counts/ num_samples) * 100
value_counts_and_percentages = zip(unique_elements, counts, percentages)

for i, j, k in value_counts_and_percentages: 
    print(f"{i}: Count {j}, Percentage: {k:.2f}%")
```

```python
bid = binom(n, p)
x = np.arange(0, n+1)

fig, axs = plt.subplots(1, 2 , figsize=(8,3))
axs[0].bar(unique_elements, percentages, color = 'blue')
axs[0].set_xlim(-1, n+1)
axs[0].set_ylim(0, 100)
axs[0].set_title("Binomial distribution (n,p) - Sample", fontsize='10')
axs[0].set_xlabel("Values of random variable x", fontsize='10')
axs[0].set_ylabel("percentage", fontsize='10')

axs[1].bar(x, bid.pmf(x), color = 'green')
axs[1].set_xlim(-1, n+1)
axs[1].set_ylim(0,1)
axs[1].set_title('Binomial distribution (n, p)- Population', fontsize='10')
axs[1].set_xlabel('Values of variable x', fontsize='10')
axs[1].set_ylabel('Probablity', fontsize='10')

plt.tight_layout()
plt.show()
```
